```python
carros <- mpg %>%     # mexemos e depois guardamos
  mutate(cambio = substr(trans, 1, 1)) %>%
  mutate(cambio = case_when(cambio == "a" ~ "Automático",
                            cambio == "m" ~ "Manual")) %>%
  mutate(tracao = case_when(drv == "f" ~ "Dianteira",
                            drv == "4" ~ "4 x 4",
                            drv == "r" ~ "Traseira")) %>%
 ## "joga fora" drv e trans
  select(-drv, -trans)
```

```python
head(carros)
```

```python
## grafico vazio
carros %>%
  ggplot()
```

```python
## eixos x e y
carros %>%
  ggplot(mapping = aes(x=cambio, y=hwy))
```

```python
## forma boxplot
carros %>%
  ggplot(mapping = aes(x=cambio, y=hwy)) +
    geom_boxplot()
```

```python
## pinta
carros %>%
  ggplot(mapping = aes(x=hwy, y=cambio)) +
    ## faz o bloxpot
    geom_boxplot(aes(fill = cambio, show.legend=FALSE)) +
    ylab("Câmbio") +
    xlab("Consumo na estrada(mpg)") +
    ggtitle("Combustível de acordo com o tipo de câmbio") +
    ## tira a tela cinza
    theme_minimal() +
    theme(plot.title = element_text(hjust=0.5))
```

```python
carros %>%
  ggplot(mapping = aes(x=hwy)) +
    geom_histogram(aes(fill=cambio), alpha = 0.4, color = "black", binwidth = 5) +
    xlab("Consumo na estrada(mpg)") +
    ylab("Frequência") +
    theme_minimal()
```

```python
carros %>%
  ggplot(mapping = aes(x=hwy)) +
    geom_histogram(aes(fill=cambio), alpha = 0.4, color = "black", binwidth = 5) +
    ## separa os graficos melhorando a visualização
    facet_grid(rows = vars(cambio)) +
    xlab("Consumo na estrada(mpg)") +
    ylab("Frequência") +
    theme_minimal()
```

```python
carros %>%
  ggplot(mapping = aes(x=hwy)) +
    ## after_stat pede para fazer conta
    geom_histogram(aes(fill=cambio, y = after_stat(density)),
    color="black", binwidth=5, show.legend=FALSE) +
    facet_grid(rows = vars(cambio)) +
    xlab("Consumo de combustível na estrada") +
    ylab("Frequência") +
    theme_minimal()
```

```python
titulos <- fread("https://datasets.imdbws.com/title.basics.tsv.gz", quote="", na.strings="\\N")
```

```python
glimpse(titulos)
```

```python
unique(titulos$titleType)
```

```python
titulos %>%
  filter(titleType == "movie", startYear <= 2024) %>%
  group_by(startYear) %>%
  summarise(Tempo = mean(runtimeMinutes, na.rm = TRUE)) %>%
  ggplot(mapping = aes(x=startYear, y =Tempo)) +
  geom_line()
```

```python
titulos %>%
  filter(titleType == "movie", startYear <= 2024, startYear >= 1910) %>%
  group_by(startYear) %>%
  summarise(Tempo = mean(runtimeMinutes, na.rm=TRUE)) %>%
  ggplot(mapping = aes(x=startYear, y=Tempo)) +
    geom_line() +
    xlab("Ano de lançamento") +
    ylab("Duração média") +
    ggtitle("Duração média de filmes ao longo do tempo") +
    theme_minimal() +
    theme(plot.title = element_text(hjust=0.5))
```

geom_col = grafico de colunas

##grafico de pizza
no `x` colocamos o fator 1 (`factor(1)`), contamos (`stat="bin"`) e no `fill` o que deve ser contado (no nosso exemplo, `sex`).
* O `geom_bar` usa `position = "fill"` para preencher todo o espaço.

geom_bar() = grafico de barras



```python
penguins %>%
  ggplot(mapping = aes(x=factor(1), stat="bin", fill=sex)) +
    geom_bar()
```

* Para terminar, colocamos `y` em coordenada polar (`coord_polar`) e despoluímos a figura com `theme_void()`
* coordenada retangular --> coordenada polar

```python
penguins %>%
  ggplot(mapping = aes(x=factor(1), stat="bin", fill=sex)) +
    geom_bar() +
    coord_polar("y", start=0) +
    theme_void()
```
