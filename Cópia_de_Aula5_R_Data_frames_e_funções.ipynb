# Criando nossas próprias funções
Podemos criar funções! A sintaxe é
```
nome_da_funcao <- function(arg1, arg2, ..., argk){
   # tarefa a ser realizada
}
```

````r
# Função que calcula a soma de dois números
minha_soma <- function(x,y){
  soma <- x+y
  soma
}
````

````r
class(minha_soma)
````

````r
minha_soma(2,3)
````

````r
minha_soma(c(1,2),c(3,4))
````

````r
c(1,2) + c(3,4)
````

````r
# Podemos usar return para explicitar o que deve ser retornado
minha_soma2 <- function(x,y){
  return(x+y)
}
````

Podemos adaptar a função `minha_soma()` para tratar erros.

````r
minha_soma <- function(x, y) {
  if (class(x) == "numeric" & class(y) == "numeric") {
     x + y
  }
}
````

````r
minha_soma(3,2)
````

````r
minha_soma("a",2)
````

## `return()` e `stop()`
* `return()` sai da função imediatamente com a resposta (dentro do `return`).
* `stop()` sai da função imediatamente e coloca mensagem de erro (dentro do `stop`).

````r
minha_soma_NA <- function(x, y) {
  if (class(x) == "numeric" & class(y) == "numeric") {
     soma <- x + y
     return(soma)
  }
  NA
}
````

````r
minha_soma_NA(1, 2)
````

````r
minha_soma_NA("a", "b")
````

````r
minha_soma_NA(1, "b")
````

````r
minha_soma_erro <- function(x, y) {
  if (class(x) != "numeric" | class(y) != "numeric") {
    stop("A classe dos objetos x e y deve ser numérica.")
  }
  x + y
}
````

````r
minha_soma_erro(1, 2)
````

````r
minha_soma_erro("a", "b")
````

````r
minha_soma_erro(1, "b")
````

## `for`
Sintaxe:
```
for(indice in vetor){
  # Faça coisas
}
```

````r
numero_de_colunas <- ncol(mtcars)

for (coluna in 1:numero_de_colunas) {
  media <- mean(mtcars[,coluna])

  print(media)
}
````

````r
numero_de_colunas <- ncol(mtcars)

# Antes, criamos um vetor vazio.
medias <- c()

for (coluna in 1:numero_de_colunas) {
  medias[coluna] <- mean(mtcars[,coluna])
}

medias
````
