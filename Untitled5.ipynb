Crie uma função que recebe três inteiros e retorna sua soma. Porém, quando algum valor é divisível por 3 o mesmo não é considerado na soma. Casos todos os três sejam divisíveis por 3 a soma deve resultar em 0.

9 %% 3 == 0 ( 9 é divisível por três = TRUE)   %%(resto)

```python
# não for divisível por 3, soma
Soma <- function(x,y,z){
  resp <- 0
    if (x %% 3 != 0)
    resp <- resp + x
    if (y %% 3 != 0)
    resp <- resp +y
    if (z %% 3 != 0)
    resp <- resp + z
    {
      return(resp)
    }
}
```

```python
Soma(13,6,9)
# não conta o 6 e o 9
```

```python
soma3 <- function(x,y,z){
    if(x %% 3 != 0){
        a <- x
    }
    else{
      a <- 0
    }
    if(y %% 3 != 0){
        b <- y
    }
    else{
      b <- 0
    }
    if(z %% 3 != 0){
        c <- z
    }
    else{
      c <- 0
    }
    return(a+b+c)
}
```

```python
soma3(5,9,2)
```

```python
somar3 <- function(x,y,z){
  return(x*(x %% 3 !=0) + y*(y %% 3 !=0) + z*(z %% 3 !=0))
}
```

```python
somar3(3,8,13)
```

```python
somageral <- function(vetor){
    return(sum(vetor*(vetor %% 3 != 0)))
}
somageral(c(4,6,12))
```
