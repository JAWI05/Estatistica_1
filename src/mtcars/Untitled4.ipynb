```r
por_cento <- function(valor,porcem){
  return(valor*porcem/100)
}
```

```r
por_cento(20,1)
#20%
```

```r
por_cento(porcem = 1, valor = 20)
```

```r
por_cento(11000,20)
#220000%
```
