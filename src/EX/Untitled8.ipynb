```python
library(tidyverse)
library(ggmap)
library(maps)
library(mapdata)
```

```python
mundo <- map_data("world")
head(mundo)
```

```python
mundo %>%
  ggplot() +
    ## desenhar poligonos
    geom_polygon(aes(x=long, y=lat, group=group, fill=region), color="black", show.legend = FALSE) +
      coord_fixed(1.3) +
      theme_minimal()
```

```python
expectativa <- read_csv("https://raw.githubusercontent.com/cyshine/EstatESEG/main/Life%20Expectancy%20Data.csv")
```

```python
expec2015 <- expectativa %>%
  ##filtrar somente 2015
  filter(Year == 2015) %>%
  rename(region = Country)
```

```python
glimpse(expec2015)
```

A função `left_join` junta dois conjuntos de dados. Por isso mudamos `Country` para `region` no conjunto `expec2015`: para poder juntar pela coluna comum `region` nos dois conjuntos `expec2015` e `mundo`.

```python
mundo2 <- mundo %>% left_join(expec2015)
```
