# O pipe %>%
Serve para encadear um monte de funções uma dentro da outra.

No R Studio, use Ctrl-Shift-M para fazer o pipe.

Aqui, só digite %>%

```python
# soma todos e tira raiz
x <- c(1,2,3,6,30)
sqrt(sum(x))
```

```python
# Tem o mesmo resultado - (pipe)
x %>% sum %>% sqrt
```

Entendeu o que aconteceu? Pega o x, faz a soma sum e depois tira a raiz quadrada sqrt.

O resultado dá no mesmo, mas quando temos que usar muitas funções usar parêntesis fica muito difícil de ler.

```python
# Selecionando só nome, tipo, tempo e gênero
head(select(titulos, primaryTitle, originalTitle, runtimeMinutes, genres))
```

```python
titulos %>%
   select(titulos, primaryTitle, originalTitle, runtimeMinutes, genres)
```

```python
# Selecionando consecutivos
select(titulos, titleType:runtimeMinutes)
```

Algumas funções úteis:
* `starts_with`: colunas cujos nomes começam com algum texto
* `ends_with`: colunas cujos nomes terminam com algum texto
* `contains`: colunas cujos nomes contêm algum padrão

```python
select(titulos, starts_with("title"))
```

```python
select(titulos, ends_with("Title"))
```

```python
select(titulos, ends_with("Title"), ends_with("Year"))
```

```python
# Fazendo os mais novos
arrange(titulos, desc(startYear))
```

```python
#Ordenar os mais velhos para os mais novos
arrange(titulos, startYear)
```

```python
# "Filtrar" os dados de acordo com um critério
filter(titulos,titleType=="videoGame")
```

```python
filter(titulos, runtimeMinutes > 60, isAdult==0)
```

```python
# Mais de um critério? lançamemtos de jogos
filter(titulos, titleType == "videoGame", startYear >= 2023)
```

```python
# Primeiro filtra os videogames de gênero ação selcionar as colunas e ordena dos mais novos para os mais velhos
titulos %>%
  filter(titleType == "videoGame", str_detect(genres,"Action")) %>%
  select(primaryTitle, startYear, genres) %>%
  arrange(desc(startYear))
```

```python
titulos %>%
filter(titleType == "videoGame", str_detect(primaryTitle, "Zelda")) %>%
select(primaryTitle, startYear, genres)
arrange(startYear)
```

```python
titulos %>%
filter(titleType == "movie", str_detect(primaryTitle, "Star Wars")) %>%
select(primaryTitle, startYear, genres, runtimeMinutes) %>%
arrange(startYear)
```

#Fazendo contas na base: mutate
Podemos usar mutate para fazer algumas contas.

```python
titulos %>%
  mutate(tempoHoras = runtimeMinutes/60)
```

```python
titulos %>%
  mutate(tempoHoras = runtimeMinutes/60) %>%
  filter(tempoHoras > 1, titleType == "movie", isAdult == 0) %>%
  select(primaryTitle, startYear, tempoHoras, genres) %>%
  arrange(desc(tempoHoras))
```

#Resumindo dados: summarise
Podemos calcular algumas estatísticas com a função summarise.

```python
#média do tempo
titulos %>%
  summarise(tempo = mean(runtimeMinutes,na.rm=TRUE))
```

```python
# Posso fazer mais resumos
titulos %>%
  summarise(media = mean(runtimeMinutes, na.rm=TRUE),
            mediana = median(runtimeMinutes, na.rm=TRUE),
            desvio = sd(runtimeMinutes, na.rm=TRUE))
```

```python
# Posso também resumir várias colunas de uma vez
titulos %>%
  summarise(media = mean(runtimeMinutes, na.rm=TRUE),
            mediana = median(runtimeMinutes, na.rm=TRUE),
            desvio = sd(runtimeMinutes, na.rm=TRUE),
            maisvelho = min(startYear, na.rm=TRUE))
```

#group_by

```python
titulos %>%
  group_by(titleType) %>%
  summarise(media = mean(runtimeMinutes, na.rm=TRUE),
            qtde = n()) %>%
  ungroup  # Em geral é boa prática retirar os grupos depois
```
