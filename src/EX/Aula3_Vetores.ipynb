## Fatiando vetores
Em Python, `v[1:7]` mostra os elementos 1 a 6 do iterável `v`. Em R?

Quase a mesma coisa. Ele mostra os elementos 1 a **7** do vetor `v`.

Também podemos colocar qualquer sequência de índices (um vetor!) dentro dos colchetes.

Veja os exemplos a seguir.

````r
v <- 11:20
v
# Fatiando v
# Quero os três primeiros elementos. Como faço?
v[1:3]
````

# Contas com vetores
O que acontece se a gente usar as operações nos vetores?

````r
salarios <- c(10000, 12000, 9000, 8500, 12000, 11000)
salarios
````

````r
salarios
salarios + 1000
````

````r
salarios
salarios+salarios*0.1
````

````r
aumentos <- c(1000, 500, 800, 1200, 1000, 800)
salarios
aumentos
salarios+aumentos
````

````r
aumentos.perc <- c(0.1, 0.2, 0.15, 0.08, 0.09, 0.12)
salarios
salarios*aumentos.perc
salarios + salarios*aumentos.perc
````

````r
# Dá para fazer mais contas!
salarios*(1+aumentos.perc)
````

## Reciclagem
Tente as seguintes contas:
* `v1+v2` com `v1 <- c(1,2,3,4)` e `v2 <- c(10,20)`
* `v1+v3` com `v1` como anteriormente e `v3 <- c(10,20,30)`

````r
v1 <- c(1,2,3,4)
v2 <- c(10,20)
v1+v2 # o mesmo que c(1,2,3,4) + c(10,20,10,20)
````

````r
v3 <- c(10,20,30)
v1+v3 # c(1,2,3,4) + c(10,20,30,10)
````

# Vetores e operadores lógicos
Operadores lógicos
* `&`: E
* `|`: OU
* '!' : não
* `==`, `!=`: igual, diferente
* `>`, `<`, `>=`, `<=`
* `%in%`: pertencer

````r
!TRUE
````

````r
p <- c(TRUE, FALSE)
q <- c(TRUE, TRUE, FALSE, FALSE)
p & q
!p
p | q
````
