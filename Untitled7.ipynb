#Encoding
Os arquivos de computador são feitos com caracteres, mas o modo que os sistemas operacionais (Windows, Linux, Android, Mac etc) codificam os caracteres pode variar. Isso é o encoding.

O Windows usa latin1, windows-1250, windows-1252, ISO-8859-2, ISO-8859-1, ...
Linux e Mac usam UTF-8.
Em geral, aparecem problemas com encoding quando há caracteres acentuados ou especiais.

```r
frase_com_acentos <- "Você comerá manga essa tarde"
```

```r
#tipo de fonte usada
Encoding(frase_com_acentos)
```

```r
# latin1 - é um padrão do windows
Encoding(frase_com_acentos) <- "latin1"
```

```r
frase_com_acentos
```

Vamos terminar de limpar: o peso e a altura são caracteres (<chr>). Vamos ajeitar com parse_number.

```r
names(pokemon)
```

```r
pokemon <- pokemon %>%
  mutate(`Weight.(Kg)` = parse_number(`Weight.(Kg)`),
         `Height.(m)` = parse_number(`Height.(m)`))
```

```r
head(pokemon)
```

```r
pokemon %>%
  group_by(Type_0) %>%
  summarise(Peso_medio = mean(`Weight.(Kg)`, na.rm=TRUE),
            Altura_medio = mean(`Height.(m)`, na.rm=TRUE))
```
