```python
## cria uma coluna nova             mutate
## case_when (toda letra a irá se tornar automático)
## select()

carros <- mpg %>%     # mexemos e depois guardamos
  mutate(cambio = substr(trans, 1, 1)) %>%
  mutate(cambio = case_when(cambio == "a" ~ "Automático",
                            cambio == "m" ~ "Manual")) %>%
  mutate(tracao = case_when(drv == "f" ~ "Dianteira",
                            drv == "4" ~ "4 x 4",
                            drv == "r" ~ "Traseira")) %>%
  select(-drv, -trans)
```

```python
carros %>%
  ggplot(mapping = aes(x = displ, y = hwy)) +
    geom_point(size = 4, alpha = 0.4) +
    facet_grid(cambio ~ tracao, scales = "free") +
    geom_smooth(method = "lm") +
    coord_cartesian() +
    scale_y_continuous(breaks = seq(0,50,5)) +
    scale_x_continuous(breaks = seq(0,10,0.5)) +
    xlab("Deslocamento do motor (volume em litros)") +
    ylab("Milhas por galão em rodovias (MPG)") +
    ggtitle("Consumo de combustível") +
    theme(axis.text = element_text(size = 10),
          axis.title = element_text(size = 20),
          axis.title.y = element_text(margin = margin(t = 0, r = 20, b = 0, l = 0)),
          axis.title.x = element_text(margin = margin(t = 20, r = 0, b = 0, l = 0)),
          plot.title = element_text(size = 30, face = "bold", hjust = 0.5,
                                    color = 'Blue',
                                    margin = margin(t = 0, r = 0, b = 20, l = 0)),
          strip.background = element_rect(colour = "black", fill = "white"),
          strip.text = element_text(face = "bold", size = 16))
```

## Camada 5: Estatísticas (opcional)
Para inserir modelos estatísticos, usamos a próxima camada.

No nosso exemplo, parece que dá para passar uma reta em cada gráfico. Para fazer isso, usamos `geom_smooth` (sim, podemos passar dois `geom` no mesmo gráfico!)

Lembre-se: use https://ggplot2.tidyverse.org/reference/ e https://livro.curso-r.com/8-1-o-pacote-ggplot2.html para ver as possibilidades!

```python
## tipo de coordenada
carros %>%
  ggplot(mapping = aes(x = displ, y = hwy)) +
    geom_point(size = 2, alpha = 0.4) +
    facet_grid(rows = vars(cambio),
               cols = vars(tracao),
               scales = "free") +
    geom_smooth(method = "lm") +
    coord_cartesian() +
    scale_y_continuous(breaks=seq(0,50,2.5)) +
    scale_x_continuous(breaks=seq(0,10,0.5)) +
    xlab("Deslocamento do motor") +
    ylab("Consumo na estrada (mpg)") +
    ggtitle("Consumo de combustível de acordo com câmbio e tração")
```
